```python
import pandas as pd
import os
```

<span style="font-family: Arial, sans-serif; font-size: 20px;">Task 1: Combine Monthly Sales Data into a Single CSV File

```python
path = 'D:\Visual code studio projects\VSC projects\JupyterProject\Database/'
frames = []
for file in os.listdir(path):
    if file.endswith(".csv"):
        filepath = path + file
        dataframe = pd.read_csv(filepath)
        frames.append(dataframe)


result = pd.concat(frames)
result.to_csv("AnnualSales2019.csv", index = False)
df = result 

```

<span style="font-family: Arial, sans-serif; font-size: 20px;"> Task 2: Adding and cleaning data

<span style="font-family: Arial, sans-serif; font-size: 14px;">Task 2.1: Remove "Nan" or "Nah" values

```python
df = df.dropna()
df.head()
```

<span style="font-family: Arial, sans-serif; font-size: 20px;">Task 3: Calculate monthly sales



<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 3.1: Adding new column 'Sales'

```python
df['Price Each'] = pd.to_numeric(df['Price Each'], errors = 'coerce')
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors = 'coerce')
df['Sales'] = df['Price Each'] * df['Quantity Ordered']
df
```

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 3.2: Calculate total sales each month

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors = 'coerce')
df
```

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 3.3: Calculate total sales in specific day and day time

```python
df = df.dropna(subset = ['Order Date'])
total_sales_by_date = df.groupby(df['Order Date'].dt.date)['Sales'].sum()
try:
    total_sales_by_month = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
except Exception as e:
    print(f"Error during aggregation: {e}")

for month, sales in total_sales_by_month.items():
    print(f"{month}: {sales:.2f}")
                     
for date, sales in total_sales_by_date.items():
    print(f"{date}: {sales:.2f}")

df.head(20)
```

<span style="font-family: Arial, sans-serif; font-size: 20px;"> Task 4: Analyze data based on locations

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 4: Add new column "City"

```python
df['City'] = df['Purchase Address'].str.split(',').str[1].str.strip()
df
```

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 4.2: Calculate total sales in each city

```python
total_sales_by_city = df.groupby(df['City'])['Sales'].sum().reset_index()
print(total_sales_by_city)
```

<span style="font-family: Arial, sans-serif; font-size: 20px;"> Task 5: Analyze best time to display ads to maximize profits.

```python
total_sales_by_time_of_day = df.groupby(df['Order Date'].dt.hour)['Sales'].sum().reset_index()
sorted_total_sales_by_time_of_day = total_sales_by_time_of_day.sort_values(by = 'Sales', ascending = False)
print(sorted_total_sales_by_time_of_day)
df.head(40)
```

<span style="font-family: Arial, sans-serif; font-size: 20px;"> Task 6: Analyze products often sold together

```python
from itertools import combinations
from collections import Counter
duplicate_id = df[df.duplicated(subset = 'Order ID')]['Order ID']
duplicate_rows = df[df['Order ID'].isin(duplicate_id)]
order_grouped_by_id = duplicate_rows.groupby('Order ID')['Product']
product_pairs = []
for order_id, products in  order_grouped_by_id:
    pair = combinations(sorted(products), 2)
    product_pairs.extend(pair)

pair_counts = Counter(product_pairs)
pair_counts = Counter.most_common(pair_counts)
for item, count in pair_counts:
    print(f"{item}: {count}")
```

<span style="font-family: Arial, sans-serif; font-size: 20px;"> Task 7: Analyze best-selling products

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 7.1: Calculate total Quantity of Products Sold in descending order

```python
total_quantity_sold  = df.groupby(['Product', 'Price Each'])['Quantity Ordered'].sum().reset_index()
total_quantity_sold = total_quantity_sold.sort_values(by = 'Quantity Ordered', ascending = False)
total_quantity_sold['Sales'] = total_quantity_sold['Price Each'] * total_quantity_sold['Quantity Ordered']
total_quantity_sold
```

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 7.2: Analyze the reasons for high and low for total quantity order(Co-occurrences counting)

```python
product_co_occurrences = Counter()
for items, count in pair_counts:
    product1, product2 = items
    product_co_occurrences[product1] = product_co_occurrences[product1] + count
    product_co_occurrences[product2] = product_co_occurrences[product2] + count
product_co_occurrences = Counter.most_common(product_co_occurrences)
for item, count in product_co_occurrences:
    print(f"{item}: {count}")
print(product_co_occurrences)
```

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 7.2: Analyze the reasons for high and low for total quantity order(Location related)

```python
best_selling_product_by_city = df.groupby(['City', 'Product'])['Quantity Ordered'].sum().reset_index()
best_selling_product_by_city = best_selling_product_by_city.sort_values(by = 'Quantity Ordered', ascending = False)
best_selling_product_by_city = best_selling_product_by_city.drop_duplicates(subset = ['City'])
best_selling_product_by_city
```

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 7.3: Analyze the reasons for high and low for total sales(Location related)

```python
most_profitable_product_by_city = df.groupby(['City', 'Product'])['Sales'].sum().reset_index()
most_profitable_product_by_city = most_profitable_product_by_city.sort_values(by = 'Sales', ascending = False)
most_profitable_product_by_city = most_profitable_product_by_city.drop_duplicates(subset = ['City'])
most_profitable_product_by_city
```

<span style="font-family: Arial, sans-serif; font-size: 14px;"> Task 7.3: Analyze the reasons for high and low for total sales(Daytime related)

```python
best_selling_product_by_time = df.groupby([df['Order Date'].dt.hour, 'Product'])['Quantity Ordered'].sum().reset_index()
best_selling_product_by_time = best_selling_product_by_time.sort_values(by = 'Quantity Ordered', ascending = False)
best_selling_product_by_time = best_selling_product_by_time.drop_duplicates(subset = ['Order Date'])
best_selling_product_by_time
```
